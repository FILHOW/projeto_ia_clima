```python
import pandas as pd
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

df1 = pd.read_csv('Weather Station Locations.csv')
df2 = pd.read_csv('Summary of Weather.csv')
```

```python
df_merged = pd.merge(df1, df2, left_on='WBAN', right_on='STA', how='inner')
```

```python
df_merged.head() 
```

```python
df_merged = df_merged.drop('STA', axis=1)
df_merged.head() 
```

```python
# Primeiro, substituímos o valor 'T' por 0.0
df_merged['Precip'] = df_merged['Precip'].replace('T', 0.0)
# Em seguida, convertemos para um tipo numérico, e os valores que ainda
# não puderem ser convertidos serão transformados em NaN.
df_merged['Precip'] = pd.to_numeric(df_merged['Precip'], errors='coerce')
# Por fim, removemos as linhas que agora contêm NaN, garantindo que
# o seu DataFrame não tenha mais valores ausentes na coluna 'Precip'.
df_merged.dropna(subset=['Precip'], inplace=True)

df_merged['Snowfall'] = df_merged['Precip'].replace('T', 0.0)
# Em seguida, convertemos para um tipo numérico, e os valores que ainda
# não puderem ser convertidos serão transformados em NaN.
df_merged['Snowfall'] = pd.to_numeric(df_merged['Precip'], errors='coerce')
# Por fim, removemos as linhas que agora contêm NaN, garantindo que
# o seu DataFrame não tenha mais valores ausentes na coluna 'Precip'.
df_merged.dropna(subset=['Snowfall'], inplace=True)
```

```python
df_merged = df_merged[["WBAN","STATE/COUNTRY ID","Precip"]]
df_merged.head()
```

```python
df_processed = pd.get_dummies(df_merged, columns=['STATE/COUNTRY ID'], dtype = int)
df_processed.head()
```

```python
features = df_processed.columns.tolist()

coluna_a_remover = 'Precip'
if coluna_a_remover in features:
    features.remove(coluna_a_remover)

coluna_a_remover2 = 'WBAN'
if coluna_a_remover2 in features:
    features.remove(coluna_a_remover2)

print(features)
```

```python
X = df_processed[features]
y = df_processed['Precip']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
modelo = LinearRegression()
modelo.fit(X_train, y_train)
y_pred = modelo.predict(X_test)
print(y_pred)
```

```python
# Calculando o MAE
mae = mean_absolute_error(y_test, y_pred)
print(f'MAE: {mae}')

# Calculando o MSE
mse = mean_squared_error(y_test, y_pred)
print(f'MSE: {mse}')

# Calculando o RMSE
rmse = np.sqrt(mse)
print(f'RMSE: {rmse}')

# Calculando o R-quadrado
r2 = r2_score(y_test, y_pred)
print(f'R-quadrado: {r2}')
```
